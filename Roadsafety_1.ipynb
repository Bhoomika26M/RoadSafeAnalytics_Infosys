# Load and Analyze the Dateset

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\acer\OneDrive\Desktop\DS_INTERN\US_Accidents_March23.csv")
df.head()

df.shape

df.columns

df.info()

df.describe()
```

# Find Null Values for Each Columns

```python
df.isnull().sum().sort_values(ascending=False)
```

# Drop columns have too many null values

```python
drop_cols = ["End_Lat", "End_Lng", "Precipitation(in)", "Wind_Chill(F)"]
df.drop(columns=drop_cols, inplace=True)
```

# Fill null values Use Median()

```python
num_cols = ["Wind_Speed(mph)","Visibility(mi)", "Humidity(%)","Temperature(F)", "Pressure(in)"]

for col in num_cols:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
df["Visibility(mi)"].describe()
df["Visibility(mi)"].value_counts().head()
```

# Fill null Values With "Unknown"

```python
cat_cols_unknown = ["Weather_Condition", "Wind_Direction",
                    "Airport_Code", "City", "Street", "Timezone", "Zipcode"]

for col in cat_cols_unknown:
    if col in df.columns:
        df[col].fillna("Unknown", inplace=True)
```

# Fill null values use Mode()

```python
twilight_cols = ["Sunrise_Sunset", "Civil_Twilight",
                 "Nautical_Twilight", "Astronomical_Twilight"]

for col in twilight_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

# fill null with Empty String

```python
df["Description"].fillna("", inplace=True)
```

# Replaces null values with Start_Time Column

```python
df["Weather_Timestamp"] = pd.to_datetime(df["Weather_Timestamp"])
df["Weather_Timestamp"].fillna(df["Start_Time"], inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

# Add New Columns as per MileStone

```python
df["Start_Time"] = pd.to_datetime(df["Start_Time"], format="mixed")

df["Hour"] = df["Start_Time"].dt.hour
df["Weekday"] = df["Start_Time"].dt.day_name()
df["Month"] = df["Start_Time"].dt.month_name()
df[["Hour", "Weekday","Month"]].head(5)
```

# Encode categorical variables as per Milestone

```python
df["Weather_Condition"] = df["Weather_Condition"].astype("category")
df["Weather_Condition"].head()
```

# Remove Outliers

```python
for col in ["Wind_Speed(mph)", "Humidity(%)", "Visibility(mi)",
            "Temperature(F)", "Pressure(in)"]:

    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)

    df[col] = df[col].clip(lower, upper)
```

# Remove Duplicates

```python
df = df.drop_duplicates()
```

#  Save Cleaned Dataset

```python
df.to_csv("US_Accidents_Cleaned_dataset.csv", index=False)
```
