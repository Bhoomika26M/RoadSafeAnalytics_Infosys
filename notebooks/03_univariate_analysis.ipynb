```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "US_Accidents_Cleaned.csv"
df = pd.read_csv(file_path, nrows=200000)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
df["Severity"].value_counts().sort_index().plot(kind="bar")
plt.title("Accident Severity Distribution")
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["Severity"].plot(kind="hist", bins=4)
plt.title("Severity Distribution Histogram")
plt.xlabel("Severity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df["Hour"].value_counts().sort_index().plot(kind="bar")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df["Weekday"].value_counts().plot(kind="bar")
plt.title("Accidents by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(10,4))
df["Month"].value_counts().plot(kind="bar")
plt.title("Accidents by Month")
plt.xlabel("Month")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(10,4))
df["Weather_Condition"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Weather Conditions During Accidents")
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(10,4))
df["Street"].value_counts().head(10).plot(kind="bar")
plt.title("Top Road Types Involved in Accidents")
plt.xlabel("Road Type")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df["Street"].value_counts().head(5).plot(kind="pie", autopct="%1.1f%%")
plt.title("Top 5 Road Types (Percentage)")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["Junction"].value_counts().plot(kind="bar")
plt.title("Accidents Near Junctions")
plt.xlabel("Junction Present")
plt.ylabel("Accident Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["Traffic_Signal"].value_counts().plot(kind="bar")
plt.title("Accidents Near Traffic Signals")
plt.xlabel("Traffic Signal Present")
plt.ylabel("Accident Count")
plt.show()
```
