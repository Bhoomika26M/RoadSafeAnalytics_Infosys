```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("US_Accidents_Cleaned.csv", nrows=200000)

# Display basic info
df.head()
```

```python
hourly_accidents = df["Hour"].value_counts().sort_index()

plt.figure(figsize=(10,5))
hourly_accidents.plot(kind="bar")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
hourly_accidents = df["Hour"].value_counts().sort_index()

plt.figure(figsize=(10,5))
hourly_accidents.plot(kind="bar")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
monthly_accidents = df["Month"].value_counts().sort_index()

plt.figure(figsize=(10,5))
monthly_accidents.plot(kind="line", marker="o")
plt.title("Accidents by Month")
plt.xlabel("Month")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
weather_severity = df.groupby("Weather_Condition")["Severity"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
weather_severity.plot(kind="bar")
plt.title("Average Accident Severity by Weather Condition")
plt.xlabel("Weather")
plt.ylabel("Average Severity")
plt.show()
```

```python
visibility_severity = df.groupby("Visibility(mi)")["Severity"].mean()

plt.figure(figsize=(10,5))
plt.scatter(visibility_severity.index, visibility_severity.values, alpha=0.6)
plt.xlabel("Visibility (miles)")
plt.ylabel("Average Severity")
plt.title("Visibility vs Accident Severity")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["Start_Lng"], y=df["Start_Lat"], alpha=0.2)
plt.title("Geographical Distribution of Accidents")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
severity_counts = df["Severity"].value_counts().sort_index()

plt.figure(figsize=(6,4))
severity_counts.plot(kind="bar")
plt.title("Distribution of Accident Severity")
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.show()
```
