```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("US_Accidents_Cleaned.csv", nrows=200000)
```

```python
hourly_accidents = df["Hour"].value_counts().sort_index()

plt.figure(figsize=(8,4))
hourly_accidents.plot(kind="bar")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Accident Count")
plt.show()
```

```python
peak_hour = hourly_accidents.idxmax()
peak_hour
```

```python
weather_severity = df[df["Weather_Condition"].isin(["Rain", "Fog"])]

plt.figure(figsize=(6,4))
sns.boxplot(
    x="Weather_Condition",
    y="Severity",
    data=weather_severity
)
plt.title("Severity During Rain vs Fog")
plt.xlabel("Weather Condition")
plt.ylabel("Severity Level")
plt.show()
```

```python
corr_value = df["Severity"].corr(df["Visibility(mi)"])
corr_value
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x="Visibility(mi)",
    y="Severity",
    data=df,
    alpha=0.3
)
plt.title("Visibility vs Severity")
plt.xlabel("Visibility (miles)")
plt.ylabel("Severity")
plt.show()
```

```python
summary = pd.DataFrame({
    "Question": [
        "Peak accident time",
        "Severity during rain vs fog",
        "Visibility vs Severity"
    ],
    "Conclusion": [
        "Accidents peak during rush hours",
        "Rain-related accidents are more severe",
        "Lower visibility increases severity"
    ]
})

summary
```
